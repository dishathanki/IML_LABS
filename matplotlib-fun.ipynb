* it allows users to create numerous and diverse plot types.
* It can be used in variety of user interfaces such as IPhython shells, Python scripts, Jupyter notebooks, as well as web applications and GUI toolkits.

# Functions of Matplotlib 

# 1. pyplot.plot()


* matplotlib.pyplot.plot() function is a fundamental component of the Matplotlib library , specially within the pyplot module.
* It provides an implicit way of plotting.
* It also opens figures on your screen, and acts as the figure GUI manager.


* Syntax: matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

* Parameters:

* x, y: These parameters represent the horizontal and vertical coordinates of the data points. 

* fmt: This is an optional parameter that contains a string value.used to specify the format of the plot.

* data: An optional parameter, ‘data’ refers to an object with labeled data.

```python
# this is an example fo line plot 

import matplotlib.pyplot as plt
import numpy as np
 
# Plotting a simple line graph
plt.plot([1, 2, 3])
plt.title('matplotlib line plot')
plt.show()
```

# 2. show()

* show() ffunction provides a unified interface for creating different types of plots.

* The simplest example uses the plot() function to plot values as x,y coordinates in a data plot.

* show() function starts an event loop, looks for all currently active figure objects, and opens one or more interactive windows that display your figure or figures.

```python
import matplotlib.pyplot as plt
import numpy as np

# X axis parameter:
xaxis = np.array([2, 8])

# Y axis parameter:
yaxis = np.array([4, 9])

plt.plot(xaxis, yaxis)
plt.show()
```

# 3. Labels: xlabel(), ylabel() 

* xlabel() and ylabel() functions to set a label for the x- and y-axis.

* it will be easy in understanding if you provide appropriate label to x and y axis !

```python
import matplotlib.pyplot as plt
import numpy as np

# X axis parameter:
xaxis = np.array([2, 8])

# Y axis parameter:
yaxis = np.array([4, 9])

plt.plot(xaxis, yaxis)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# 4. Grid: grid()

* grid() function of axes object sets visibility of grid inside the figure to on or off.

* You can also display major / minor (or both) ticks of the grid.

* Additionally color, linestyle and linewidth properties can be set in the grid() function.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95])
y = np.array([240, 250, 260, 270])

plt.plot(x, y)
plt.grid()
plt.show()
```

# 5. Bars: bar()

* The bars are positioned at x with the given alignment. 

* Their dimensions are given by height and width. 

* The vertical baseline is bottom (default 0).

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

# 6. Histogram: hist()

* A histogram is a graph showing frequency distributions.

* It is a graph showing the number of observations within each given interval

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

# 7. Subplot: subplot()

* With the subplot() function you can draw multiple plots in one figure.

* it takes three arguments that describes the layout of the figure.

* layout is organized in rows and columns, which are represented by the first and second argument.

* The third argument represents the index of the current plot.

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
# it means 2 row 1 column and first plot 
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
# it means 2 row 1 column and second plot 
plt.plot(x,y)

plt.show()
```

# 8. pie chart: pie()

* you can use the pie() function to draw pie charts.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

* As you can see the pie chart draws one piece (called a wedge) for each value in the array.

* By default the plotting of the first wedge starts from the x-axis and moves counterclockwise.


**example :**

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

# 9. Save the plotted images into pdf: savefig()

* As the name suggests savefig() method is used to save the figure created after plotting data.

* The figure created can be saved to our local machines by using this method.

* Filename .png for image.

**parameters :**


* fname = Filename .png for image, .pdf for pdf format. File location can also be specified here.

* format = File format such as .png, .pdf.

* facecolor and edgecolor	Default as White.

* Orientation	Landscape or Portrait.

```python
# importing required modules  
import matplotlib.pyplot as plt 
  
# creating plotting data 
xaxis =[1, 4, 9, 16] 
yaxis =[1, 2, 3, 4] 
  
# plotting  
plt.plot(xaxis, yaxis) 

# Make sure you use savefig() before show(). 
plt.savefig("squares.png") 
  
plt.show()
```
