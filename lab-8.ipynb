# 1. Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

```python
# Name : Disha Thanki
# Enrollment Number : 216270307024

# importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# reading csv file
data = pd.read_csv('/kaggle/input/stock-market-data-nifty-50-2000-2022/Nifty_50_with_indicators_.csv')
data.head()

# setting data column as the index of dataset
data.set_index('Date', inplace=True)

# specifing starting and ending date
start_date = '09 Jun 2022'
end_date = '03 Jun 2022'

# Filtering data that include data between specified date only 
filtered_data = data[start_date:end_date]

# Create a line plot of opening and closing stock prices
plt.plot(filtered_data.index, filtered_data['open'], label='Open Price', marker='o')
plt.plot(filtered_data.index, filtered_data['close'], label='Close Price', marker='o')

# setting plot and lables and title 
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices between {} and {}'.format(start_date, end_date))

#ploting graph 
plt.legend()
plt.show()
```
